```python
import os
import json
import requests

# ThÆ° má»¥c lÆ°u dá»¯ liá»‡u
save_dir = "data_scraper/raw_data"
os.makedirs(save_dir, exist_ok=True)

# Base URL API
base_url = "https://api-crownx.winmart.vn/plg/api/web/item/collection?pageNumber={}"

# Láº·p qua tá»« trang 1 Ä‘áº¿n 20
for page in range(1, 21):
    url = base_url.format(page)
    print(f"ðŸ”¹ Äang láº¥y dá»¯ liá»‡u tá»« trang {page}...")

    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        if response.status_code == 200:
            data = response.json()

            # LÆ°u tá»«ng trang thÃ nh file riÃªng
            file_path = os.path.join(save_dir, f"page_{page}.json")
            with open(file_path, "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=2)

            print(f"âœ… ÄÃ£ lÆ°u: {file_path}")
        else:
            print(f"âš ï¸ Lá»—i {response.status_code} táº¡i trang {page}: {response.text[:100]}")
    except Exception as e:
        print(f"âŒ Lá»—i táº¡i trang {page}: {e}")

print("\nðŸŽ‰ HoÃ n táº¥t! Táº¥t cáº£ dá»¯ liá»‡u Ä‘Ã£ Ä‘Æ°á»£c lÆ°u vÃ o thÆ° má»¥c:", save_dir)
```
